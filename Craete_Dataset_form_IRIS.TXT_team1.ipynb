# Import and load dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing as prep
from scipy.stats import variation
# :D
df = pd.read_csv('Data/iris.txt', header=None)
df.info()
df
```

# Change header

```python
df.columns = ['sepalLength', 'sepalWidth', 'petalLength', 'petalWidth', 'species']
df.head(3)
```

# Find and drop duplicate training example

```python
df.drop_duplicates(inplace=True)
df.info()
```

# Encoding species

```python
labelEncoder = prep.LabelEncoder()
df['species'] = labelEncoder.fit_transform(df['species'])
df['species'].unique()
```

# Mean, SDT, min, max, and quarter

```python
df.describe()
```

# Variance and CV

```python
df.var()
```

```python
variation(df, axis=0)
```

# Covariance and Correlation

```python
df.cov()
```

```python
df.corr()
```

# Total variance

```python
totalVar = 0
for i in range(4):
    totalVar += np.cov(df.T)[i][i]
print(f'Total variance is : {totalVar}')
```

# Plot dataframe

```python
sns.pairplot(df, hue='species')
plt.show()
```

```python
sns.violinplot(x='species', y='sepalLength', data=df)
plt.show()
sns.violinplot(x='species', y='sepalWidth', data=df)
plt.show()
sns.violinplot(x='species', y='petalLength', data=df)
plt.show()
sns.violinplot(x='species', y='petalWidth', data=df)
plt.show()
```
