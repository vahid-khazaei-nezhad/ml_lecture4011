# Get year from guidance

```python
guidance = df['guidance']
year = []
for i in range(0, 2530):
    try:
        if guidance[i][-5] == '،':
            year.append(guidance[i][-4:])
    except:
        pass
# convert 1378 to ۱۳۷۸
for i in range(0, len(year)):
    if year[i] == "1378":
        year[i] = "۱۳۷۸"
yeardf = pd.DataFrame(year)
yeardf
```

# Find frequancy of years

```python
unique, frequency = np.unique(year, return_index=True)
count = np.asarray((unique, frequency))
yearFreq = pd.DataFrame(count).T
yearFreq.head()
```

# Drop some rows

```python
yearFreq.drop([0, 1, 47, 48, 49, 50], inplace=True)
yearFreq.head()
```

# Change header

```python
yearFreq.columns = ['year', 'frequency']
yearFreq.head()
```
